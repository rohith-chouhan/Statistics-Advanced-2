### 1) What is hypothesis testing in statistics?
#### > Hypothesis testing in statistics is a formal procedure used to make decisions or inferences about a population based on sample data.It helps us determine whether there is enough evidence to support a particular claim (called a hypothesis) about the population.

### 2) What is the null hypothesis, and how does it differ from the alternative hypothesis?
#### > **Null Hypothesis (H₀)**:
- It is the default assumption in hypothesis testing.
- States that there is no effect, no difference, or no relationship in the population.
- It represents the "status quo" or baseline claim.
#### > **Alternative Hypothesis (H₁ or Ha)**:
- It is what we want to test and possibly prove with evidence.
- States that there is an effect, a difference, or a relationship.
- It is accepted only if sample evidence strongly contradicts H₀.

### 3) Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
#### > The significance level, denoted by α, is the threshold probability that helps us decide whether to reject the null hypothesis (H₀).It represents the risk of making a Type I error - rejecting H₀ when it is actually true.
#### **Role in Hypothesis Testing**:
- Set the cutoff point for evidence against H₀.
- If the p-value ≤ α - reject H₀ (evidence supports H₁).
- If the p-value > α - fail to reject H₀ (not enough evidence against H₀).
- Controls the risk of error:α = 0.05 means we are willing to accept a 5% chance of wrongly rejecting H₀.

### 4) What are Type I and Type II errors? Give examples of each.
#### > **Type I Error (False Positive)**:
- Occurs when we reject the null hypothesis (H₀) even though it is true.
- In other words: we think there is an effect/difference, but in reality, there isn't.
- Probability of making a Type I error = α (significance level).
##### **Example:**
- A medical test for a disease:
- H₀: The patient does not have the disease.
- H₁: The patient has the disease.
- If the test result wrongly says the patient has the disease (when they don't), it's a Type I error.
- Real-world impact: A healthy person is diagnosed as sick (false alarm).
#### > **Type II Error (False Negative)**:
- Occurs when we fail to reject the null hypothesis (H₀) even though it is false.
- In other words: we think there is no effect/difference, but in reality, there is.
- Probability of making a Type II error = β (beta).
##### **Example**:
- Same medical test:
- H₀: The patient does not have the disease.
- H₁: The patient has the disease.
- If the test result wrongly says the patient is healthy (when they actually have the disease), it's a Type II error.
- Real-world impact: A sick person goes undiagnosed (missed detection).

### 5)What is the difference between a Z-test and a T-test? Explain when to use each.
#### > **Z-test**: A hypothesis test that uses the Z-distribution (standard normal distribution).
##### **When to use**:
- Large sample size (n ≥ 30)
- Population standard deviation (σ) is known
##### **Example**:
- Testing whether the average height of students in a large school is 160 cm, given the population variance is known.
#### > **T-test**: A hypothesis test that uses the T-distribution (similar to normal but with heavier tails).
##### **When to use**:
- Small sample size (n < 30)
- Population standard deviation (σ) is unknown (we estimate it using sample standard deviation s)
##### **Example**:
- Testing whether the average marks of 15 students differ from 50 when the population variance is unknown.

### 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
#### > A Z-score tells us how many standard deviations a data point is away from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = [(x - mean) / std for x in data]
    return z_scores

data = [45, 47, 50, 52, 53, 49, 48, 46, 51, 54]
z_scores = calculate_z_scores(data)

print(f"Original Data: {data}")
print(f"Z-scores: {[f'{z:.2f}' for z in z_scores]}")

plt.hist(z_scores, bins=5, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.show()
```
